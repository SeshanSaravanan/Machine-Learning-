## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting 

%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize = (10, 10)) #(width, height)

# 3. Plot data 
ax.plot(x, y)

# 4. Customize Plot 
ax.set(title = "Simple Plot",
       xlabel= "x-axis",
       ylabel= "y-axis")

# 5. Save figure and show 
fig.savefig("images/sample-plot.png")
```

### Two options for plotting subplots 

```python
# Option 1 

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                             ncols = 2,
                                             figsize = (10,5))

# Plot to each different axis

ax1.plot(x, x/2); 
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Option 2

fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize = (10,5))

# Plot to each different axis

ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

# Which one to use? (Pyplot vs matplotlib OO method)

*When plotting something quickly, use pyplot method

*When plotting something more advanced, use OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot method - plotting directly using python 

over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target");
```

```python
# OO mehod mixed with pyplot method

fig, ax = plt.subplots(figsize = (10, 6)) # size follows width, height
over_50.plot(kind="scatter",
             x = "age",
             y = "chol",
             c = "target",
             ax = ax);
```

```python
## OO method from Scratch

fig, ax = plt.subplots(figsize = (10,6))

# Plot the data 
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"])

#Customize the plot 
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend - basically the info in the top right corner of the graph 
ax.legend(*scatter.legend_elements(), title= "target") 

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(),
           linestyle = "--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize=(10,10),
                               sharex = True)

# Add data to ax0
scatter = ax0.scatter(x= over_50["age"],
                     y= over_50["chol"],
                     c= over_50["target"])

# Customize the ax0
ax0.set(title = "Heart Disease and Cholestrol Levels",
        ylabel = "Cholestrol");

# Add legend to ax0 
ax0.legend(*scatter.legend_elements(), title = "Target");


# Add a meanline or horizontal line 
ax0.axhline(y= over_50["chol"].mean(),
            linestyle = "--");


## Add data to ax1
scatter = ax1.scatter(x= over_50["age"],
                     y= over_50["thalach"],
                     c= over_50["target"])

# Customize the ax1
ax1.set(title = "Heart Disease and Max Heart Rate Levels",
        xlabel = "Age",
        ylabel = "Max Heart Rate");

# Add legend to ax1 
ax1.legend(*scatter.legend_elements(), title = "Target");


# Add a meanline or horizontal line 
ax1.axhline(y= over_50["thalach"].mean(),
            linestyle = "--");

## Add a title to the figure 

fig.suptitle("Heart Disease Analysis", fontsize = 18, fontweight = "bold");
```

```python
# Set the style 
plt.style.use("seaborn-whitegrid")

## OO method from Scratch

fig, ax = plt.subplots(figsize = (10,6))

# Plot the data 
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap = "winter") #changes the color scheme

#Customize the plot 
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

# Add a legend - basically the info in the top right corner of the graph 
ax.legend(*scatter.legend_elements(), title= "target") 

# Add a horizontal line 
ax.axhline(over_50["chol"].mean(),
           linestyle = "--");
```

```python
# Customizing the x and y axis limitations 
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize=(10,10),
                               sharex = True)

# Add data to ax0
scatter = ax0.scatter(x= over_50["age"],
                     y= over_50["chol"],
                     c= over_50["target"],
                     cmap="winter")

#Change the x - axis limits
ax0.set_xlim([50, 80])

# Customize the ax0
ax0.set(title = "Heart Disease and Cholestrol Levels",
        ylabel = "Cholestrol");

# Add legend to ax0 
ax0.legend(*scatter.legend_elements(), title = "Target");


# Add a meanline or horizontal line 
ax0.axhline(y= over_50["chol"].mean(),
            linestyle = "--");


## Add data to ax1
scatter = ax1.scatter(x= over_50["age"],
                     y= over_50["thalach"],
                     c= over_50["target"],
                     cmap = "winter")
#Change the x - axis and y-axis limits 
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Customize the ax1
ax1.set(title = "Heart Disease and Max Heart Rate Levels",
        xlabel = "Age",
        ylabel = "Max Heart Rate");

# Add legend to ax1 
ax1.legend(*scatter.legend_elements(), title = "Target");


# Add a meanline or horizontal line 
ax1.axhline(y= over_50["thalach"].mean(),
            linestyle = "--");

## Add a title to the figure 

fig.suptitle("Heart Disease Analysis", fontsize = 18, fontweight = "bold");
```
